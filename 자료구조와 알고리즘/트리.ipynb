### 트리의 정의
* 한개 이상의 노드로 이루어진 유한집합
* 최상위 노드의 자식 노드들은 n개의 분리 집합으로 분리될 수 있다.
* 말단의 노드는 잎 노드라고 한다.

### 용어 정리
노드 - 트리의 원소

간선 - 노드를 연결하는 선, `부모노드와 자식노드를 연결` -> 부모와 자식을 연결하는 노드라는 것이 중요!

루트노드 - 트리의 시작 노드

형제 노드 - 같은 부모노드의 자식 노드 (촌수 문제 나오겠다)

조상노드 - 간선을 따라 루트노드까지 이르는 경로에 있는 모든 노드 -> 조상의 형제는 조상이 아님에 유의!

자손 노드 - 서브트리에 있는 하위 레벨의 노드들 -> 대상 노드 B를 루트로 하는 노드들

노드의 차수 - 노드에 연결된 자식 노드의 수 -> 손자노드는 제외, only 자식노드임

노드의 높이 - 루트에서 노드에 이르는 간선의 수, 노드의 레벨

트리의 높이 - 트리에 있는 노드의 높이 중에서 가장 큰 값. 최대 레벨

> 모든 노드가 2개의 서브트리를 갖는 특별한 형태의 트리

* 이진 트리의 최대 차수는 2이다.

### 포화 이진 트리
* 모든 레벨에 노드가 포화상태로 차 있는 이진트리
* 높이가 h일 때 모든 노드의 개수는 2**(h+1) -1의 노드를 가지고 있다.
* 포화 이진트리와 완전 이진트리만 루트를 1번으로 고정한다.

### 완전 이진 트리

* 순서가 빈틈없이 채워져 있기 때문에 1차원 배열로 표현하기 편하다
* 완전 이진트리의 범주 안에 포화이진트리가 포함된다.
* 우선순위 큐를 구현 할 때 완전이진트리를 사용한다.

```python

def pre_order(T):
  if T > 0:
    print(T, end=' ')
    pre_order(left[T])
    pre_order(right[T])


def in_order(T):
  if T > 0:
    in_order(left[T])
    print(T, end=" ")
    in_order(right[T])


def post_order(T):
  if T > 0:
    post_order(left[T])
    post_order(right[T])
    print(T, end=" ")

V = int(input())
E = V-1
arr = list(map(int, input().split()))

left = [0]* (V+1)
right = [0]* (V+1)
par = [0]* (V+1)

for i in range(E):
  p, c = arr[i*2], arr[i*2+1]
  par[c] = p
  if left[p] == 0:
    left[p] = c
  else:
    right[p] = c

root =1
for i in range(1, V+1):
  if par[i]== 0:  # 부모노드가 없는 경우
    root = i
    break

pre_order(root)
```
