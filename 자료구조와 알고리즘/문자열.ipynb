# 코드 체계

```python
# 회문
def ispalindrome(txt):
    for i in range(len(txt)//2):
        if txt[i] != txt[len(txt)-i]
            return False
    return True
```

```python
# 사전 순서 비교 - 유니코드를 서로 비교한다
def my_strcmp(s1,s2):
    if s1 < s2:
        return -1
    elif s1 > s2:
        return 1
    else:
        return 0
```
