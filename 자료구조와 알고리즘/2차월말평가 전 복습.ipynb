```python
# 찾는 값이 배열에 있으면 해당 원소의 인덱스, 없으면 -1을 idx에넣기
lst = [0,1,2,3,4,5]
N = len(lst)
val = 3
idx = -1
for i in range(N):
    if lst[i] == val:
        idx = i
        break
print(val)
```

```python
# 버블정렬

lst = [5,4,3,2,1]
def bubble_sort(lst):
    N = len(lst)
    for i in range(N-1,0, -1): # i번쨰 정렬마다 정렬횟수가 한번씩 줄어듬
        for j in range(i): # 왜 i번이 아니라 i-1번 정렬하지?
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    return lst
print(bubble_sort(lst))
# 군데 이렇게 하면 거꾸로 정렬되어있지 않은 배열도 정렬이 되나?
```

```python
# 카운팅 정렬
def counting_sort(lst):
    N = len(lst)
    cnt_arr = [0] * (max(lst) + 1)
    temp = [0]*N
    for i in range(N): # N번 계산
        cnt_arr[lst[i]] += 1

    for i in range(1, max(lst)+1): # 여기서 k번 개
        cnt_arr[i] += cnt_arr[i-1]

    # lst[i] = 정렬해야 할 원소
    # cnt_arr[lst[i]] -= 1 삽입해야 할 인덱스 구하
    for i in range(N-1, -1, -1): # 여기서 N번 계산
        cnt_arr[lst[i]] -= 1
        temp[cnt_arr[lst[i]]] = lst[i] 
    return temp

print (counting_sort(lst))
```

```python
# 순차검색 while문 버전
arr = [1,2,3,4,5,6,]
def sequential_search(arr, length, key):
    idx = 0
    while idx<length and arr[idx] != key:
        idx += 1
    if idx<length: return idx
    else: return -1

# 찾는 값보다 크면 검색 안하기
def sequential_search2(arr, length, key):
    idx = 0
    while idx<length and arr[idx] < key:
        idx += 1

    # 찾는 값이 없으면 -1을 반환하기 위한 if
    if i<length and arr[idx] == key:
        return idx
    else:
        return -1

print(sequential_search(arr, len(arr), 6))
print(sequential_search2(arr, len(arr), 6))
```

```python
# 이진검색
arr = [1,2,3,4,5,6,]
def binary_search(arr, key):
    start = 0
    end = len(arr)-1
    while start<=end:
        middle = (start + end)//2
        if arr[middle] == key:
            return middle
        elif arr[middle] >key:
            end = middle -1
        else: 
            start = middle +1

    return -1
binary_search(arr, 3)
```

```python
# 선택 정렬
arr = [7,4,4,7,8,4,3,5,1,3,45,8,64]
def selection_sort(arr):
    N = len(arr)
    for i in range(N-1):
        min_idx = i
        for j in range(i+1, N):
            if  arr[min_idx]> arr[j]:
                min_idx = j
        arr[min_idx], arr[i] = arr[i], arr[min_idx]
    return arr
        
            
print(selection_sort(arr))
```

```python
# 셀렉션 알고리즘
def selection(arr, k):
    N = len(arr)
    for i in range(k):

        # 최솟값의 인덱스 찾기
        min_idx = i
        for j in range(i+1, N):
            if  arr[min_idx]> arr[j]:
                min_idx = j
        arr[min_idx], arr[i] = arr[i], arr[min_idx]
    return arr[k-1]
        
            
print(selection(arr, 8))
```
