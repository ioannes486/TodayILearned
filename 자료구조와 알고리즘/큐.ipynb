# 큐 자료구조
> 먼저 들어온 데이터가 먼저 나가는 선형 자료구조
스택과 마찬가지로 삽입과 삭제의 위치가 제한적인 자료구조로 큐의 뒤에서는 삽입만 하고 앞에서는 삭제만 이루어진다

### 선입선출
가장 먼저 넣은 자료가 가장 먼저 나오는 것
서비스 대기행렬이 있음

### 큐의 연산 과정
front -> 꺼내기 : popleft의 대상이 되는 원소의 인덱스
rear -> 넣기 : popright의 대상이 되는 원소의 인덱스

front == rear : 스택이 비워진 상태임

### 선형 큐
배열이나 연결형 리스트로 구현할 수 있음
큐의 크기는 배열의 크기와 같다
front : 가장 최근에 삭제됭 원소의 인덱스
rear : 마지막으로 저장된 원소의 인덱스

### 상태표현
초기 상태 : front == rear == -1

공백상태 front == rear

꽉 찬 상태 : rear == n-1

# 원형 큐
### 선형 큐 이용시의 문제점
> 잘못된 포화상태의 인식 : 앞에 공간이 남아있음에도 rear가 ㅜ-1이면 꽉 찬걸로 인식한다

### 원형 큐
> 공간낭비를 막기 위해 처음과 끝이 연결된 구조

### 원형 큐의 구조
* 초기 공백 상태 : front = rear = 0
* 인덱스의 순환 : front와 rear가 마지막 인덱스로 이동하면 나머지 연산을 사용하여 다시 0으로 이동하는 것이 핵심
* front 변수 ; 공백상태와 포화상태 구분을 쉽게 하기 위해 프론트가 있는 자리는 사용하지 않고 항상 빈자리로 둔다. > 값을 저장하지 않고 항상 0으로 둔다
* 얘 왜 빈자리로 둠?

### 원형 큐의 구현



```python
# my_queue

front = rear = -1

q = [0]* 10

rear += 1
q[rear] = 1
rear += 1
q[rear] = 2
rear += 1
q[rear] = 3

# 빼기
front += 1
print(q[front])
front += 1
print(q[front])
front += 1
print(q[front])
def is_empty():

    return front == rear

def is_full():

    return (rear + 1) % len(q) == front  # 맨 마지막이 될  인덱스가 맨 처음 인덱스와 같으면

def enqueue(elem):
    if is_full():
        print("Queue Full")
    else:
        rear += (rear + 1) % len(q)
        q[rear] = elem
        
def dequeue():
    if is_empty():
        print("Queue Empty")
    else:
        front += 1
    return q[front]

# 검색 : 가장 앞에 있는 원소를 반환하는 연산, front 앞에 있는 원소를 반환
# 근데 프론트가 왜 맨 앞의 원소가 아니라 그 전의 원소에 인덱스가 있지? -> front를 하나 증가시킨 후 그 원소를 꺼내는 연산을 하기 위함
def qpeek():
    if is_empty():
        print("Queue Empty")
    else:
        return q[front+1]
```

### deque
> 컨테이너 자료형 중 하나로 양쪽 끝에서 빠르게 추가와 삭제를 할 수 있는 리스트류 컨테이너, 연결 리스트를 직접 만들지 않아도 된다.

### deque의 연산
* `append(x)` : 오른쪽에 원소 추가
* `popleft(x)` : 왼쪽에서 원소 뺴기

```python
# 연결 큐 구현하기 : 빠르게 처리할 수 있는 큐
class Node :
    def _init_(self, item, n=None):
        self.item = item
        self.next = n

def enqueue(item): # 연결 큐의 삽입 연산
    global front, rear
    newNode = Node(item) # 새로운 노드 생성
    if front == None : # 큐가 비어있다면
        front = newNode
    else :
        rear.next = newNode
        rear = newNode


def is_empty() :
    return front == None

def dequeue() : # 연결 큐의 삭제 연산
    global front, rear
    if is_empty() :
        print("Queue_Empty")
        return None

    item = front.item
    front = front.r
    if front == None :
        rear = None
    return item


def qpeek() :
    return front.item

def print_q() :
    f = front
    s = ""
    while f :
        s += f.item + " "
        f = f.next
    return s

front = None
rear = None
```
